```python
#calculate avg ratings
avg_ratings=df['Ratings'].mean()
print(f'Average Ratings:{avg_ratings:.2f}')
```

```python
#distribution of ratings
plt.figure(figsize=(8,6))
sns.histplot(df['Ratings'],kde=True,bins=10)
plt.title("Distribution of ratings")
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
#bar plot of avg discount by brand
plt.figure(figsize=(8,6))
sns.barplot(x='Brand',y='Discount',data=df,estimator=np.mean)
plt.title("Avg Discount By Brand")
plt.xlabel('Brand')
plt.ylabel('Avg Discount(%')
plt.xticks(rotation=45)
plt.show()
```

```python
#distribution of ratings
plt.figure(figsize=(8,6))
sns.countplot(x='Brand',data=df)
plt.title("Number of Products By Brand")
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
#Average Ratings for Top 5 Brands by Count
top_brands=df['Brand'].value_counts().head(5).index
plt.figure(figsize=(10,6))
sns.barplot(x='Brand',y='Ratings',data=df[df['Brand'].isin(top_brands)],estimator=np.mean)
plt.title('Average ratings for top 5 brands')
plt.xlabel('Brand')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
#Convert MSP column to numeric(this will convert invalid parsing to NaN)
df['MSP']=pd.to_numeric(df['MSP'],errors='coerce') #coerce replace empty fields by 'NaN'
#Now drop rows with NaN values in MSP column if needed
df=df.dropna(subset=['MSP'])
#define price range segments
bins=[0,10000,20000,30000,50000,100000]#define ranges
labels=['Low-End(<10K)','Budget(10K-20K)','Mid-Range(20K-30K)','Premium(30K-50K)','Flagship(>50K)']
```

```python
#create new column for price range
df['Price_Range']=pd.cut(df['MSP'],bins=bins,labels=labels)
#Count of mobiles in each price range
price_range_count=df['Price_Range'].value_counts()
plt.figure(figsize=(8,6))
sns.barplot(x=price_range_count.index,y=price_range_count.values)
plt.title("Mobile count in different price segments")
plt.xlabel('Price Range')
plt.ylabel('Number of Mobiles')

plt.tight_layout()
plt.show
```

```python
brand_highest_discount=df.groupby('Brand')['Discount'].max().reset_index()
top_brands_by_discount=brand_highest_discount.sort_values(by='Discount',ascending=False).head(5)
plt.figure(figsize=(10,6))
sns.barplot(x='Brand',y='Discount',data=top_brands_by_discount,palette='coolwarm',hue='Brand',legend=False)
plt.title("Top 5 brands with highest discount")
plt.xlabel('Brand')
plt.ylabel('Discount Amount')
plt.tight_layout()
plt.show()
```

```python
df.groupby('Brand')['Ratings'].mean().nlargest(5).plot(kind='bar')
plt.title('Top 5 Brands by Avg ratings')
plt.ylabel('Average Ratings')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
top_5_brands_by_reviews=df.groupby('Brand')['No_of_reviews'].sum().sort_values(ascending=False).head(5)
top_5_brands_by_reviews_df=top_5_brands_by_reviews.reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='Brand',y='No_of_reviews',data=top_5_brands_by_reviews_df,palette='viridis',hue='Brand',legend=False)
plt.title("Top 5 brands with highest discount")
plt.title('Top 5 brands by total no of reviews')
plt.xlabel('Brand')
plt.ylabel('Total No of Reviews')
plt.tight_layout()
plt.show()
```
